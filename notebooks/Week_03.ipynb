## Part 1: Introduction to Loops


Loops are one of the most fundamental and powerful concepts in programming. They allow you to execute a block of code **multiple times** without writing it repeatedly. Think of loops as automation tools â€“ they let you write code once and run it as many times as needed.

### Why Do We Need Loops?

Imagine you need to print numbers from 1 to 1000. Without loops, you would need to write 1000 print statements! With loops, you can do it in just 2 lines of code.

- **Automation** â€“ Write code once, run it many times
- **Processing collections** â€“ Work through lists, strings, files, databases
- **Calculations** â€“ Sum numbers, find averages, accumulate results
- **Pattern generation** â€“ Create sequences, shapes, and complex structures
- **Data analysis** â€“ Process sensor readings, log files, user inputs

### Types of Loops in Python

| Loop Type | Best Used When | Example Use Case |
| --- | --- | --- |
| `for` | Number of iterations is **known** | Processing all items in a list |
| `while` | Number of iterations is **unknown** | Input validation, waiting for condition |

> ğŸ’¡ **Note:** **Rule of Thumb:** If you know how many times you need to repeat, use `for`. If you need to repeat until something happens, use `while`.

#### Loop vs. Manual Repetition

**Figure 1.1: Loop vs Manual**

````python
# WITHOUT loops (bad practice):
print("Hello 1")
print("Hello 2")
print("Hello 3")

print("\n--- With a loop (good practice) ---\n")

# WITH loops (good practice):
for i in range(1, 4):
    print(f"Hello {i}")

# The loop version is shorter, cleaner, and easier to modify!
````

## Part 2: The for Loop


The `for` loop is the most common loop in Python. It iterates over any **iterable** object â€“ strings, lists, tuples, ranges, dictionaries, and more.

### Basic Syntax

```
for variable in iterable:
    # Code to execute for each item
    # (indented with 4 spaces)
```

#### Simple for Loop Examples

**Figure 2.1: Basic for Loop**

````python
# Iterate over a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}")

print("\n--- Iterating over a string ---\n")

# Iterate over a string
word = "Python"
for letter in word:
    print(letter, end=" ")
````

#### Sum of Numbers

**Figure 2.2: Sum Calculation**

````python
# Calculate sum of numbers
numbers = [10, 20, 30, 40, 50]
total = 0

for num in numbers:
    total += num
    print(f"Added {num}, running total: {total}")

print(f"\nFinal sum: {total}")
print(f"Average: {total / len(numbers)}")
````

## Part 3: The range() Function


The `range()` function generates a sequence of numbers. It's commonly used with `for` loops when you need to iterate a specific number of times.

### range() Syntax

| Syntax | Description | Example |
| --- | --- | --- |
| `range(stop)` | 0 to stop-1 | `range(5)` â†’ 0,1,2,3,4 |
| `range(start, stop)` | start to stop-1 | `range(2, 6)` â†’ 2,3,4,5 |

#### range() Examples

**Figure 3.1: range() Variations**

````python
# range(stop) - 0 to stop-1
print("range(5):", list(range(5)))

# range(start, stop) - start to stop-1
print("range(2, 7):", list(range(2, 7)))

# range(start, stop, step) - with step
print("range(0, 10, 2):", list(range(0, 10, 2)))

# Negative step (counting down)
print("range(10, 0, -1):", list(range(10, 0, -1)))
````

## Part 4: Iterating Over Collections


Python's `for` loop can iterate over any iterable object including strings, lists, tuples, sets, and dictionaries.

#### String Processing

**Figure 4.1: String Iteration**

````python
# Count vowels in a string
text = "Hello World"
vowels = "aeiouAEIOU"
vowel_count = 0

for char in text:
    if char in vowels:
        vowel_count += 1

print(f"Text: '{text}'")
print(f"Vowel count: {vowel_count}")

# Reverse a string
reversed_text = ""
for char in text:
    reversed_text = char + reversed_text
print(f"Reversed: '{reversed_text}'")
````

#### List Operations

**Figure 4.2: List Processing**

````python
# Find min and max manually
numbers = [34, 12, 89, 45, 67, 23]

min_val = numbers[0]
max_val = numbers[0]

for num in numbers:
    if num < min_val:
        min_val = num
    if num > max_val:
        max_val = num

print(f"Numbers: {numbers}")
print(f"Minimum: {min_val}")
print(f"Maximum: {max_val}")

# Filter even numbers
evens = []
for num in numbers:
    if num % 2 == 0:
        evens.append(num)
print(f"Even numbers: {evens}")
````

#### Dictionary Iteration

**Figure 4.3: Dictionary Iteration**

````python
# Iterate over dictionary
student_scores = {
    "Alice": 92,
    "Bob": 85,
    "Charlie": 78
}

# Keys only
print("Students:")
for name in student_scores:
    print(f"  - {name}")

# Keys and values
print("\nScores:")
for name, score in student_scores.items():
    print(f"  {name}: {score}")
````

## Part 5: enumerate() and zip() Functions


`enumerate()` gives you index numbers automatically, while `zip()` lets you iterate over multiple sequences in parallel.

#### Using enumerate()

**Figure 5.1: enumerate() Basics**

````python
# enumerate() provides index + value
fruits = ["apple", "banana", "cherry"]

print("With enumerate():")
for index, fruit in enumerate(fruits):
    print(f"  {index}: {fruit}")

print("\nWith enumerate(start=1):")
for rank, fruit in enumerate(fruits, start=1):
    print(f"  #{rank}: {fruit}")
````

#### Using zip()

**Figure 5.2: zip() Basics**

````python
# zip() combines multiple lists
names = ["Alice", "Bob", "Charlie"]
scores = [92, 87, 95]

print("ğŸ“Š Student Scores:")
for name, score in zip(names, scores):
    print(f"  {name}: {score}")

# Three lists together
subjects = ["Math", "Science", "English"]
grades = ["A", "B+", "A-"]
print("\nGrade Report:")
for name, subject, grade in zip(names, subjects, grades):
    print(f"  {name} got {grade} in {subject}")
````

#### Temperature Converter

**Figure 5.3: Temperature Converter**

````python
# Convert Celsius to Fahrenheit
celsius = [0, 10, 20, 30, 40]

print("#   Celsius   Fahrenheit")
print("-" * 25)
for i, c in enumerate(celsius, start=1):
    f = c * 9/5 + 32
    print(f"{i}   {c:5.1f}Â°C â†’ {f:6.1f}Â°F")
````

## Part 6: The while Loop


The `while` loop repeats as long as a condition is `True`. Use it when you don't know in advance how many iterations are needed.

### Basic Syntax

```
while condition:
    # Code to execute while condition is True
    # Must include something that changes the condition!
```

> ğŸ’¡ **Note:** **Infinite Loop Warning!** Always ensure the condition will eventually become `False`, or your program will run forever.

#### Basic while Loop

**Figure 6.1: Basic while**

````python
# Count to 5 using while
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1  # Don't forget this!

print("Done counting!")
````

#### Sum Until Threshold

**Figure 6.2: Sum Until Threshold**

````python
# Add numbers until sum exceeds 100
total = 0
number = 1

while total <= 100:
    total += number
    print(f"Added {number}, total = {total}")
    number += 1

print(f"\nFirst sum over 100: {total}")
print(f"Numbers added: {number - 1}")
````

#### Guess the Number Game

**Figure 6.3: Number Guessing**

````python
# Simple guessing game simulation
secret = 7
guesses = [3, 9, 5, 7]  # Simulated guesses
attempts = 0

for guess in guesses:
    attempts += 1
    print(f"Guess {attempts}: {guess}")
    
    if guess == secret:
        print(f"ğŸ‰ Correct! Found in {attempts} attempts!")
        break
    elif guess < secret:
        print("  Too low!")
    else:
        print("  Too high!")
````

#### while-else Clause

**Figure 6.4: while-else**

````python
# while-else: else runs if loop completes normally
n = 5
while n > 0:
    print(f"n = {n}")
    n -= 1
else:
    print("Loop completed normally!")

print("\n--- With break (else won't run) ---")
n = 5
while n > 0:
    if n == 3:
        print("Breaking at n=3")
        break
    print(f"n = {n}")
    n -= 1
else:
    print("This won't print!")
````

## Part 7: The break Statement


The `break` statement immediately exits the loop, skipping any remaining iterations. It's useful for stopping a loop when a condition is met.

#### Find First Match

**Figure 7.1: Find First Match**

````python
# Find first even number
numbers = [7, 3, 9, 4, 2, 8]

for num in numbers:
    print(f"Checking {num}...")
    if num % 2 == 0:
        print(f"âœ“ Found first even: {num}")
        break

print("Search complete")
````

#### Prime Number Check

**Figure 7.2: Prime Check**

````python
# Check if a number is prime
num = 17
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            print(f"{num} is divisible by {i}")
            break

if is_prime:
    print(f"{num} is prime!")
else:
    print(f"{num} is not prime")
````

## Part 8: The continue Statement


The `continue` statement skips the rest of the current iteration and moves to the next one. Unlike `break`, it doesn't exit the loop entirely.

#### Skip Negative Numbers

**Figure 8.1: Skip with continue**

````python
# Sum only positive numbers
numbers = [10, -5, 20, -3, 15, -8, 25]
total = 0

for num in numbers:
    if num < 0:
        print(f"Skipping {num} (negative)")
        continue
    total += num
    print(f"Added {num}, total = {total}")

print(f"\nSum of positives: {total}")
````

#### Print Odd Numbers Only

**Figure 8.2: Filter with continue**

````python
# Print only odd numbers from 1-10
print("Odd numbers 1-10:")
for i in range(1, 11):
    if i % 2 == 0:  # Skip even numbers
        continue
    print(i, end=" ")
print()  # New line
````

#### break vs continue Comparison

**Figure 8.3: break vs continue**

````python
# Compare break and continue
print("With CONTINUE (skip 3):")
for i in range(1, 6):
    if i == 3:
        continue
    print(i, end=" ")

print("\n\nWith BREAK (stop at 3):")
for i in range(1, 6):
    if i == 3:
        break
    print(i, end=" ")
print()
````

## Part 9: Nested Loops


A nested loop is a loop inside another loop. The inner loop completes all its iterations for each iteration of the outer loop.

#### Multiplication Table

**Figure 9.1: Multiplication Table**

````python
# Print multiplication table
print("    ", end="")
for j in range(1, 6):
    print(f"{j:4}", end="")
print("\n" + "-" * 25)

for i in range(1, 6):
    print(f"{i} |", end="")
    for j in range(1, 6):
        print(f"{i*j:4}", end="")
    print()
````

#### Pattern Printing

**Figure 9.2: Star Pattern**

````python
# Right triangle pattern
rows = 5

print("Right Triangle:")
for i in range(1, rows + 1):
    for j in range(i):
        print("*", end="")
    print()

print("\nPyramid:")
for i in range(1, rows + 1):
    # Spaces
    for j in range(rows - i):
        print(" ", end="")
    # Stars
    for k in range(2 * i - 1):
        print("*", end="")
    print()
````

#### Matrix Operations

**Figure 9.3: Matrix Sum**

````python
# 2D list (matrix) operations
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matrix:")
for row in matrix:
    print(row)

# Sum all elements
total = 0
for row in matrix:
    for element in row:
        total += element

print(f"\nSum of all elements: {total}")
````

## Part 10: Common Loop Patterns


Here are frequently used loop patterns that you'll encounter in programming.

#### Accumulator Pattern

**Figure 10.1: Accumulator**

````python
# Accumulator pattern: building up a result
numbers = [4, 2, 8, 5, 1]

# Sum
total = 0
for num in numbers:
    total += num
print(f"Sum: {total}")

# Product
product = 1
for num in numbers:
    product *= num
print(f"Product: {product}")

# String building
chars = ['H', 'e', 'l', 'l', 'o']
word = ""
for char in chars:
    word += char
print(f"Word: {word}")
````

#### Counter Pattern

**Figure 10.2: Counter**

````python
# Count occurrences
text = "mississippi"
target = "s"
count = 0

for char in text:
    if char == target:
        count += 1

print(f"'{target}' appears {count} times in '{text}'")
````

#### Search Pattern

**Figure 10.3: Search**

````python
# Linear search
numbers = [15, 42, 7, 23, 36, 9]
target = 23
found_index = -1

for i, num in enumerate(numbers):
    if num == target:
        found_index = i
        break

if found_index != -1:
    print(f"Found {target} at index {found_index}")
else:
    print(f"{target} not found")
````

#### Filter Pattern

**Figure 10.4: Filter**

````python
# Filter elements matching criteria
scores = [45, 92, 73, 88, 56, 81, 67, 95]
passing_score = 70

passed = []
failed = []

for score in scores:
    if score >= passing_score:
        passed.append(score)
    else:
        failed.append(score)

print(f"All scores: {scores}")
print(f"Passed (â‰¥{passing_score}): {passed}")
print(f"Failed (<{passing_score}): {failed}")
````

## Part 11: Engineering Applications


#### Sensor Data Analysis

**Figure 11.1: Sensor Analysis**

````python
# Analyze temperature sensor readings
readings = [22.5, 23.1, 25.8, 24.2, 28.5, 22.0, 26.3]
threshold = 26.0

print("ğŸŒ¡ï¸ Temperature Analysis")
print("=" * 30)

# Calculate statistics
total = 0
max_temp = readings[0]
min_temp = readings[0]
alerts = 0

for temp in readings:
    total += temp
    if temp > max_temp:
        max_temp = temp
    if temp < min_temp:
        min_temp = temp
    if temp > threshold:
        alerts += 1

avg = total / len(readings)

print(f"Readings: {len(readings)}")
print(f"Average: {avg:.1f}Â°C")
print(f"Maximum: {max_temp}Â°C")
print(f"Minimum: {min_temp}Â°C")
print(f"Threshold alerts: {alerts}")
````

#### Factorial Calculation

**Figure 11.2: Factorial**

````python
# Calculate factorial: n! = n Ã— (n-1) Ã— ... Ã— 1
n = 5
factorial = 1

for i in range(1, n + 1):
    factorial *= i
    print(f"{i}! = {factorial}")

print(f"\n{n}! = {factorial}")
````

#### Fibonacci Sequence

**Figure 11.3: Fibonacci**

````python
# Generate Fibonacci sequence
n_terms = 10
a, b = 0, 1

print(f"Fibonacci sequence ({n_terms} terms):")
for i in range(n_terms):
    print(a, end=" ")
    a, b = b, a + b
print()
````

## âŒ Common Mistakes to Avoid


These are the most frequent errors students make with loops. Recognizing these patterns will save you hours of debugging!

**Infinite `while` loops**

                    Forgetting to update the loop variable: `while x < 10:` without `x += 1` inside runs forever! Always ensure the condition will eventually become `False`.

**Off-by-one with `range()`**

`range(1, 10)` gives 1â€“9, NOT 1â€“10. The end value is *exclusive*. For 1 through 10, use `range(1, 11)`.

**Modifying a list while iterating over it**

`for item in my_list: my_list.remove(item)` â†’ Skips elements! Iterate over a copy instead: `for item in my_list[:]:`

**Wrong placement of accumulator initialization**

                    Placing `total = 0` INSIDE the loop resets it every iteration. Always initialize accumulators BEFORE the loop starts.

**Confusing `break` and `continue`**

`break` exits the loop entirely. `continue` skips to the next iteration. Mixing them up changes program behavior dramatically.
