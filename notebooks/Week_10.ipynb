## Part 1: Introduction to OOP


Object-Oriented Programming (OOP) is a programming paradigm that organizes code around **objects** that combine data (attributes) and behavior (methods). Instead of writing separate functions and variables scattered throughout your program, OOP bundles related data and functionality together into cohesive units.

| Concept | Description | Real-World Analogy | Python Example |
| --- | --- | --- | --- |
| **Class** | Blueprint/template for creating objects | Blueprint for a house | `class Sensor:` |
| **Object** | Instance of a class (actual entity) | An actual house built from blueprint | `sensor1 = Sensor()` |
| **Attribute** | Data stored in an object | House color, number of rooms | `sensor.value = 25.5` |
| **Method** | Function belonging to an object | Open door, turn on lights | `sensor.read()` |

### Why Object-Oriented Programming?

Before OOP, programmers used procedural programming where data and functions were separate. This worked for small programs but became difficult to manage as programs grew larger. OOP addresses these challenges by providing better organization, reusability, and modeling of real-world concepts.

**Figure 1.1: Procedural vs OOP Approach**

````python
# PROCEDURAL APPROACH - Data and functions are separate
# Data stored in dictionaries
sensor1 = {
    'id': 'TEMP_001',
    'type': 'temperature',
    'value': 25.5,
    'unit': 'celsius'
}

sensor2 = {
    'id': 'TEMP_002',
    'type': 'temperature',
    'value': 26.3,
    'unit': 'celsius'
}

# Functions that work with sensor dictionaries
def read_sensor(sensor):
    return f"{sensor['id']}: {sensor['value']} {sensor['unit']}"

def update_sensor(sensor, new_value):
    sensor['value'] = new_value
    return sensor

# Using procedural approach
print("=== Procedural Approach ===")
print(read_sensor(sensor1))
update_sensor(sensor1, 27.0)
print(read_sensor(sensor1))
````

**Figure 1.2: OOP Approach - Data and Behavior Together**

````python
# OOP APPROACH - Data and functions bundled together
class Sensor:
    """A class representing a sensor device."""
    
    def __init__(self, sensor_id, sensor_type, value, unit):
        """Initialize sensor with its properties."""
        self.id = sensor_id
        self.type = sensor_type
        self.value = value
        self.unit = unit
    
    def read(self):
        """Return a formatted reading string."""
        return f"{self.id}: {self.value} {self.unit}"
    
    def update(self, new_value):
        """Update the sensor value."""
        self.value = new_value

# Create objects (instances) of the Sensor class
sensor1 = Sensor('TEMP_001', 'temperature', 25.5, 'celsius')
sensor2 = Sensor('TEMP_002', 'temperature', 26.3, 'celsius')

# Using OOP approach - much cleaner!
print("=== OOP Approach ===")
print(sensor1.read())
sensor1.update(27.0)
print(sensor1.read())
print(sensor2.read())
````

### Four Pillars of OOP

**Figure 1.3: OOP Core Principles**

````python
# The Four Pillars of Object-Oriented Programming
pillars = {
    "Encapsulation": {
        "description": "Bundling data and methods that operate on that data within a single unit (class)",
        "benefit": "Hides internal details, reduces complexity",
        "example": "Sensor hides how it stores/validates data internally"
    },
    "Abstraction": {
        "description": "Showing only essential features while hiding implementation details",
        "benefit": "Simplifies interaction with complex systems",
        "example": "sensor.read() - user doesn't need to know how reading works"
    },
    "Inheritance": {
        "description": "Creating new classes based on existing classes",
        "benefit": "Code reuse, hierarchical organization",
        "example": "TemperatureSensor inherits from Sensor (Week 11)"
    },
    "Polymorphism": {
        "description": "Objects of different classes can be treated uniformly",
        "benefit": "Flexibility, extensibility",
        "example": "Different sensors with same read() interface (Week 11)"
    }
}

print("=== The Four Pillars of OOP ===\n")
for i, (pillar, info) in enumerate(pillars.items(), 1):
    print(f"{i}. {pillar}")
    print(f"   Definition: {info['description']}")
    print(f"   Benefit: {info['benefit']}")
    print(f"   Example: {info['example']}")
    print()
````

A class is defined using the `class` keyword followed by the class name. By convention, class names use PascalCase (each word capitalized). Objects are created by calling the class like a function.


### Adding Methods to Classes

**Figure 2.3: Class with Methods**

````python
# Class with methods
class Dog:
    """A class representing a dog."""
    
    def bark(self):
        """Make the dog bark."""
        print("Woof! Woof!")
    
    def sit(self):
        """Make the dog sit."""
        print("*sits down*")
    
    def speak(self, message):
        """Make the dog speak with a message."""
        print(f"Dog says: {message}")

# Create a dog instance
buddy = Dog()
buddy.name = "Buddy"  # Add attribute

# Call methods
print(f"{buddy.name}:")
buddy.bark()
buddy.sit()
buddy.speak("I want treats!")
````

## Part 3: The __init__ Constructor


The `__init__` method (pronounced "dunder init") is a special method that initializes a new object. It's called automatically when you create an instance. This is where you typically set up initial attribute values.

```
class ClassName:
    def __init__(self, param1, param2):
        self.attribute1 = param1
        self.attribute2 = param2

# Creating an instance calls __init__ automatically
obj = ClassName(value1, value2)
```

### Basic Constructor

**Figure 3.1: Using __init__ Constructor**

````python
# Using the __init__ constructor
class Student:
    """A class representing a student."""
    
    def __init__(self, name, student_id, major):
        """Initialize student with name, ID, and major."""
        self.name = name
        self.student_id = student_id
        self.major = major
        self.courses = []  # Initialize with empty list
        self.gpa = 0.0     # Default value
    
    def display_info(self):
        """Display student information."""
        print(f"Student: {self.name}")
        print(f"ID: {self.student_id}")
        print(f"Major: {self.major}")
        print(f"GPA: {self.gpa}")
        print(f"Courses: {', '.join(self.courses) if self.courses else 'None'}")

# Create student instances
student1 = Student("Ali Yƒ±lmaz", "2024001", "Mechatronics")
student2 = Student("Ay≈üe Kaya", "2024002", "Computer Engineering")

# Add some courses
student1.courses = ["Python", "Electronics", "Mechanics"]
student1.gpa = 3.5

student2.courses = ["Data Structures", "Algorithms"]
student2.gpa = 3.8

# Display information
print("=== Student 1 ===")
student1.display_info()
print("\n=== Student 2 ===")
student2.display_info()
````

### Constructor with Default Parameters

**Figure 3.2: Constructor with Defaults**

````python
# Constructor with default parameter values
class Rectangle:
    """A class representing a rectangle."""
    
    def __init__(self, width=1, height=1, color="white"):
        """
        Initialize rectangle with dimensions and color.
        
        Args:
            width: Rectangle width (default 1)
            height: Rectangle height (default 1)
            color: Fill color (default "white")
        """
        self.width = width
        self.height = height
        self.color = color
    
    def area(self):
        """Calculate and return the area."""
        return self.width * self.height
    
    def perimeter(self):
        """Calculate and return the perimeter."""
        return 2 * (self.width + self.height)
    
    def describe(self):
        """Return a description of the rectangle."""
        return f"{self.color} rectangle: {self.width}x{self.height}"

# Create rectangles with different parameters
rect1 = Rectangle()                    # All defaults
rect2 = Rectangle(5, 3)               # Custom size, default color
rect3 = Rectangle(10, 5, "blue")      # All custom
rect4 = Rectangle(width=4, color="red")  # Named arguments

# Display each rectangle
rectangles = [rect1, rect2, rect3, rect4]
for i, rect in enumerate(rectangles, 1):
    print(f"Rectangle {i}: {rect.describe()}")
    print(f"  Area: {rect.area()}, Perimeter: {rect.perimeter()}")
    print()
````

#### Engineering Example: Sensor Class

**Figure 3.3: Complete Sensor Class**

````python
# Engineering: Complete sensor class with validation
class Sensor:
    """
    A class representing a sensor device.
    
    Attributes:
        sensor_id: Unique identifier for the sensor
        sensor_type: Type of measurement (temperature, pressure, etc.)
        unit: Measurement unit (celsius, hPa, etc.)
        value: Current reading value
        min_value: Minimum valid value
        max_value: Maximum valid value
    """
    
    def __init__(self, sensor_id, sensor_type, unit, 
                 min_value=-float('inf'), max_value=float('inf')):
        """Initialize sensor with validation range."""
        self.sensor_id = sensor_id
        self.sensor_type = sensor_type
        self.unit = unit
        self.min_value = min_value
        self.max_value = max_value
        self._value = None  # Private attribute
        self._reading_count = 0
    
    def read(self, value):
        """Record a sensor reading with validation."""
        if value < self.min_value or value > self.max_value:
            print(f"Warning: {value} out of range [{self.min_value}, {self.max_value}]")
            return False
        self._value = value
        self._reading_count += 1
        return True
    
    def get_value(self):
        """Get the current value."""
        return self._value
    
    def get_status(self):
        """Return sensor status information."""
        return {
            'id': self.sensor_id,
            'type': self.sensor_type,
            'value': self._value,
            'unit': self.unit,
            'readings': self._reading_count
        }

# Create different types of sensors
temp_sensor = Sensor("TEMP_001", "temperature", "¬∞C", -40, 125)
pressure_sensor = Sensor("PRES_001", "pressure", "hPa", 800, 1200)

# Simulate readings
print("=== Temperature Sensor ===")
temp_sensor.read(25.5)
print(f"Status: {temp_sensor.get_status()}")
temp_sensor.read(200)  # Out of range!
print(f"Status: {temp_sensor.get_status()}")

print("\n=== Pressure Sensor ===")
pressure_sensor.read(1013.25)
pressure_sensor.read(1015.0)
print(f"Status: {pressure_sensor.get_status()}")
````

## Part 4: Instance Methods


Instance methods are functions defined inside a class that operate on individual objects. They always receive `self` as their first parameter, which refers to the specific instance calling the method.

### Defining and Using Instance Methods

**Figure 4.1: Instance Methods**

````python
# Instance methods work with object data via self
class BankAccount:
    """A simple bank account class."""
    
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.transactions = []
    
    def deposit(self, amount):
        """Add money to the account."""
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"+${amount}")
            return True
        return False
    
    def withdraw(self, amount):
        """Remove money from the account."""
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.transactions.append(f"-${amount}")
            return True
        return False
    
    def get_balance(self):
        """Return current balance."""
        return self.balance
    
    def get_statement(self):
        """Generate account statement."""
        print(f"=== Account Statement ===")
        print(f"Owner: {self.owner}")
        print(f"Current Balance: ${self.balance}")
        print(f"Transactions: {', '.join(self.transactions)}")

# Create account and perform operations
account = BankAccount("Ali", 100)
account.deposit(50)
account.deposit(30)
account.withdraw(25)
account.get_statement()
````

### Methods That Return self

**Figure 4.2: Method Chaining**

````python
# Methods that return self enable method chaining
class QueryBuilder:
    """Build database-like queries with method chaining."""
    
    def __init__(self, table):
        self.table = table
        self.conditions = []
        self.selected_columns = "*"
        self.order_column = None
    
    def select(self, columns):
        """Select specific columns."""
        self.selected_columns = columns
        return self  # Return self for chaining
    
    def where(self, condition):
        """Add a WHERE condition."""
        self.conditions.append(condition)
        return self
    
    def order_by(self, column):
        """Add ORDER BY clause."""
        self.order_column = column
        return self
    
    def build(self):
        """Build the final query string."""
        query = f"SELECT {self.selected_columns} FROM {self.table}"
        if self.conditions:
            query += " WHERE " + " AND ".join(self.conditions)
        if self.order_column:
            query += f" ORDER BY {self.order_column}"
        return query

# Without chaining (verbose)
query1 = QueryBuilder("users")
query1.select("name, email")
query1.where("age > 18")
print("Query 1:", query1.build())

# With chaining (elegant)
query2 = (QueryBuilder("products")
          .select("name, price")
          .where("category = 'electronics'")
          .where("price < 1000")
          .order_by("price")
          .build())
print("Query 2:", query2)
````

## Part 5: Instance vs Class Attributes


Python distinguishes between **instance attributes** (unique to each object) and **class attributes** (shared by all instances). Understanding this difference is crucial for proper OOP design.

| Attribute Type | Definition | Scope | Access |
| --- | --- | --- | --- |
| **Instance Attribute** | Defined in __init__ with self | Unique to each object | self.name or obj.name |
| **Class Attribute** | Defined in class body | Shared by all objects | ClassName.name or self.name |

### Class Attributes

**Figure 5.1: Class vs Instance Attributes**

````python
# Class attribute vs instance attribute
class Employee:
    """A class representing an employee."""
    
    # Class attributes - shared by all instances
    company = "TechCorp Inc."
    employee_count = 0
    
    def __init__(self, name, department, salary):
        # Instance attributes - unique to each object
        self.name = name
        self.department = department
        self.salary = salary
        self.employee_id = Employee.employee_count + 1
        
        # Increment class attribute
        Employee.employee_count += 1
    
    def display(self):
        print(f"{self.employee_id}: {self.name} - {self.department} @ {self.company}")

# Create employees
emp1 = Employee("Ali", "Engineering", 50000)
emp2 = Employee("Ay≈üe", "Marketing", 45000)
emp3 = Employee("Mehmet", "Engineering", 55000)

# Display all employees
print("=== All Employees ===")
emp1.display()
emp2.display()
emp3.display()

# Class attribute is shared
print(f"\nTotal employees: {Employee.employee_count}")
print(f"Company name: {Employee.company}")

# Accessing class attribute via instance
print(f"emp1's company: {emp1.company}")
print(f"emp2's company: {emp2.company}")
````

### Modifying Class Attributes

**Figure 5.2: Class Attribute Modification**

````python
# Modifying class attributes
class Counter:
    """Demonstrate class attribute modification."""
    count = 0  # Class attribute
    
    def __init__(self, name):
        self.name = name  # Instance attribute
        Counter.count += 1  # Modify via class name
    
    def show(self):
        print(f"{self.name}: total count = {Counter.count}")

# Create instances
c1 = Counter("First")
c1.show()

c2 = Counter("Second")
c1.show()  # Both see the updated count
c2.show()

c3 = Counter("Third")

# All instances share the class attribute
print(f"\nAccessing via class: {Counter.count}")
print(f"Via c1: {c1.count}")
print(f"Via c2: {c2.count}")
print(f"Via c3: {c3.count}")
````

#### Engineering Example: Sensor Registry

**Figure 5.3: Sensor Registry Pattern**

````python
# Engineering: Sensor registry using class attributes
class Sensor:
    """Sensor class with automatic registration."""
    
    # Class attributes for registry
    _registry = {}
    total_sensors = 0
    
    def __init__(self, sensor_id, sensor_type):
        self.sensor_id = sensor_id
        self.sensor_type = sensor_type
        self.value = None
        
        # Register this sensor
        Sensor._registry[sensor_id] = self
        Sensor.total_sensors += 1
    
    def read(self, value):
        self.value = value
    
    @classmethod
    def get_sensor(cls, sensor_id):
        """Retrieve a sensor by ID."""
        return cls._registry.get(sensor_id)
    
    @classmethod
    def list_sensors(cls):
        """List all registered sensors."""
        print(f"=== Sensor Registry ({cls.total_sensors} sensors) ===")
        for sid, sensor in cls._registry.items():
            print(f"  {sid}: {sensor.sensor_type} = {sensor.value}")

# Create sensors
temp1 = Sensor("TEMP_001", "temperature")
temp2 = Sensor("TEMP_002", "temperature")
pres1 = Sensor("PRES_001", "pressure")

# Take readings
temp1.read(25.5)
temp2.read(26.0)
pres1.read(1013.25)

# List all sensors
Sensor.list_sensors()

# Retrieve specific sensor
sensor = Sensor.get_sensor("TEMP_001")
print(f"\nRetrieved: {sensor.sensor_id} = {sensor.value}")
````

> üí° **Note:** Be careful when assigning to class attributes via an instance (e.g., `obj.class_attr = value`). This creates a new instance attribute that shadows the class attribute, rather than modifying the shared class attribute. Always use `ClassName.attr = value` to modify class attributes.

## Part 6: Class Methods and Static Methods


In addition to instance methods, Python classes can have **class methods** (work with class data) and **static methods** (utility functions that don't need class or instance data).

| Method Type | Decorator | First Parameter | Use Case |
| --- | --- | --- | --- |
| **Instance Method** | None | `self` | Work with instance data |
| **Class Method** | `@classmethod` | `cls` | Work with class data, alternative constructors |
| **Static Method** | `@staticmethod` | None | Utility functions, don't need class/instance |

### Class Methods

**Figure 6.1: Class Methods with @classmethod**

````python
# Class methods - work with class data, use cls parameter
class Date:
    """A class representing a date."""
    
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day
    
    def __str__(self):
        return f"{self.year}-{self.month:02d}-{self.day:02d}"
    
    @classmethod
    def from_string(cls, date_string):
        """Alternative constructor: create Date from string."""
        year, month, day = map(int, date_string.split('-'))
        return cls(year, month, day)  # cls is the Date class
    
    @classmethod
    def today(cls):
        """Alternative constructor: create today's date."""
        # Simplified - just return a fixed date for demo
        return cls(2026, 2, 3)

# Create dates using different constructors
date1 = Date(2024, 12, 25)
date2 = Date.from_string("2024-06-15")
date3 = Date.today()

print(f"Date 1 (standard): {date1}")
print(f"Date 2 (from string): {date2}")
print(f"Date 3 (today): {date3}")
````

### Static Methods

**Figure 6.2: Static Methods with @staticmethod**

````python
# Static methods - utility functions, don't need self or cls
class MathUtils:
    """A collection of math utility functions."""
    
    @staticmethod
    def is_positive(number):
        """Check if a number is positive."""
        return number > 0
    
    @staticmethod
    def is_even(number):
        """Check if a number is even."""
        return number % 2 == 0
    
    @staticmethod
    def clamp(value, min_val, max_val):
        """Clamp a value between min and max."""
        return max(min_val, min(value, max_val))
    
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        """Convert Celsius to Fahrenheit."""
        return (celsius * 9/5) + 32
    
    @staticmethod
    def fahrenheit_to_celsius(fahrenheit):
        """Convert Fahrenheit to Celsius."""
        return (fahrenheit - 32) * 5/9

# Use static methods - no instance needed!
print("=== Static Methods Demo ===")
print(f"Is 5 positive? {MathUtils.is_positive(5)}")
print(f"Is 4 even? {MathUtils.is_even(4)}")
print(f"Clamp 150 to [0, 100]: {MathUtils.clamp(150, 0, 100)}")
print(f"25¬∞C in Fahrenheit: {MathUtils.celsius_to_fahrenheit(25)}¬∞F")
print(f"77¬∞F in Celsius: {MathUtils.fahrenheit_to_celsius(77):.1f}¬∞C")
````

### Combining All Method Types

**Figure 6.3: All Method Types in One Class**

````python
# Complete example using all method types
class Temperature:
    """Temperature class demonstrating all method types."""
    
    # Class attribute
    scale = "Celsius"
    
    def __init__(self, value):
        """Instance method (constructor)."""
        self.value = value
    
    # Instance method
    def display(self):
        """Display the temperature."""
        return f"{self.value}¬∞ {Temperature.scale}"
    
    # Instance method
    def is_freezing(self):
        """Check if temperature is below freezing."""
        return self.value < 0
    
    # Class method - alternative constructor
    @classmethod
    def from_fahrenheit(cls, fahrenheit):
        """Create Temperature from Fahrenheit value."""
        celsius = (fahrenheit - 32) * 5/9
        return cls(celsius)
    
    # Class method - modify class state
    @classmethod
    def set_scale(cls, new_scale):
        """Change the display scale."""
        cls.scale = new_scale
    
    # Static method - utility function
    @staticmethod
    def convert(value, from_scale, to_scale):
        """Convert between temperature scales."""
        if from_scale == to_scale:
            return value
        if from_scale == "Celsius" and to_scale == "Fahrenheit":
            return (value * 9/5) + 32
        if from_scale == "Fahrenheit" and to_scale == "Celsius":
            return (value - 32) * 5/9
        return value

# Demo all method types
temp1 = Temperature(25)  # Regular constructor
temp2 = Temperature.from_fahrenheit(77)  # Class method constructor

print("=== Temperature Demo ===")
print(f"temp1: {temp1.display()}")
print(f"temp2: {temp2.display()}")
print(f"temp1 freezing? {temp1.is_freezing()}")

# Use static method
converted = Temperature.convert(100, "Celsius", "Fahrenheit")
print(f"\n100¬∞C = {converted}¬∞F")

# Use class method to change scale
Temperature.set_scale("Kelvin")
print(f"\nAfter scale change:")
print(f"Scale is now: {Temperature.scale}")
````

### String Representation: __str__ and __repr__

**Figure 7.1: __str__ and __repr__ Methods**

````python
# __str__ and __repr__ for string representation
class Point:
    """A 2D point class."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        """Human-readable string (for print, str())."""
        return f"Point at ({self.x}, {self.y})"
    
    def __repr__(self):
        """Developer-readable string (for debugging, repr())."""
        return f"Point(x={self.x}, y={self.y})"

p = Point(3, 4)

# __str__ is used by print() and str()
print("Using print():", p)
print("Using str():", str(p))

# __repr__ is used in debugging and containers
print("Using repr():", repr(p))
print("In a list:", [p])

# Without these methods, you'd see something like:
# <__main__.Point object at 0x...>
````

### Comparison Methods

**Figure 7.2: Comparison Special Methods**

````python
# Comparison methods for custom objects
class Student:
    """Student class with comparison by grade."""
    
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    
    def __str__(self):
        return f"{self.name}: {self.grade}"
    
    def __eq__(self, other):
        """Equal: =="""
        return self.grade == other.grade
    
    def __ne__(self, other):
        """Not equal: !="""
        return self.grade != other.grade
    
    def __lt__(self, other):
        """Less than: <"""
        return self.grade < other.grade
    
    def __le__(self, other):
        """Less than or equal: <="""
        return self.grade <= other.grade
    
    def __gt__(self, other):
        """Greater than: >"""
        return self.grade > other.grade
    
    def __ge__(self, other):
        """Greater than or equal: >="""
        return self.grade >= other.grade

# Create students
s1 = Student("Ali", 85)
s2 = Student("Ay≈üe", 90)
s3 = Student("Mehmet", 85)

# Compare students
print("=== Comparing Students ===")
print(f"s1 == s3: {s1 == s3}")  # Same grade
print(f"s1 == s2: {s1 == s2}")  # Different grade
print(f"s1 < s2: {s1 < s2}")
print(f"s2 > s1: {s2 > s1}")

# Now we can sort students!
students = [s1, s2, s3]
students.sort()
print(f"\nSorted by grade: {[str(s) for s in students]}")
````

### Arithmetic Methods

**Figure 7.3: Arithmetic Special Methods**

````python
# Arithmetic methods for mathematical operations
class Vector:
    """A 2D vector class with arithmetic operations."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __add__(self, other):
        """Vector addition: +"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """Vector subtraction: -"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """Scalar multiplication: *"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __rmul__(self, scalar):
        """Right multiplication: scalar * vector"""
        return self.__mul__(scalar)
    
    def __neg__(self):
        """Negation: -vector"""
        return Vector(-self.x, -self.y)
    
    def __abs__(self):
        """Magnitude: abs(vector)"""
        return (self.x**2 + self.y**2) ** 0.5

# Create vectors
v1 = Vector(3, 4)
v2 = Vector(1, 2)

print("=== Vector Arithmetic ===")
print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v1 + v2}")
print(f"v1 - v2 = {v1 - v2}")
print(f"v1 * 2 = {v1 * 2}")
print(f"3 * v2 = {3 * v2}")
print(f"-v1 = {-v1}")
print(f"|v1| = {abs(v1)}")
````

## Part 8: Properties and Encapsulation


Encapsulation is the practice of hiding internal details and controlling access to data. Python uses **properties** to create managed attributes with getters, setters, and deleters.

### Python Naming Conventions

| Convention | Meaning | Access Level |
| --- | --- | --- |
| `name` | Public attribute | Free access from anywhere |
| `_name` | Protected (convention) | Internal use only (can still access) |
| `__name` | Private (name mangling) | Harder to access (becomes _ClassName__name) |

### Using @property Decorator

**Figure 8.1: Properties for Controlled Access**

````python
# Using @property for controlled attribute access
class Circle:
    """Circle with computed properties."""
    
    def __init__(self, radius):
        self._radius = radius  # Protected attribute
    
    @property
    def radius(self):
        """Getter for radius."""
        return self._radius
    
    @radius.setter
    def radius(self, value):
        """Setter with validation."""
        if value < 0:
            raise ValueError("Radius cannot be negative")
        self._radius = value
    
    @property
    def diameter(self):
        """Computed property - read only."""
        return self._radius * 2
    
    @property
    def area(self):
        """Computed property - read only."""
        import math
        return math.pi * self._radius ** 2
    
    @property
    def circumference(self):
        """Computed property - read only."""
        import math
        return 2 * math.pi * self._radius

# Create circle and use properties
circle = Circle(5)

print("=== Circle Properties ===")
print(f"Radius: {circle.radius}")
print(f"Diameter: {circle.diameter}")
print(f"Area: {circle.area:.2f}")
print(f"Circumference: {circle.circumference:.2f}")

# Change radius - triggers setter
circle.radius = 10
print(f"\nAfter changing radius to 10:")
print(f"Diameter: {circle.diameter}")
print(f"Area: {circle.area:.2f}")

# Try invalid value
try:
    circle.radius = -5
except ValueError as e:
    print(f"\nError: {e}")
````

#### Engineering Example: Validated Sensor

**Figure 8.2: Sensor with Property Validation**

````python
# Engineering: Sensor with validated properties
class TemperatureSensor:
    """Temperature sensor with validated properties."""
    
    def __init__(self, sensor_id, min_temp=-40, max_temp=125):
        self._sensor_id = sensor_id
        self._min_temp = min_temp
        self._max_temp = max_temp
        self._value = None
        self._status = "idle"
    
    @property
    def sensor_id(self):
        """Read-only sensor ID."""
        return self._sensor_id
    
    @property
    def value(self):
        """Current temperature value."""
        return self._value
    
    @value.setter
    def value(self, temp):
        """Set temperature with validation."""
        if temp < self._min_temp or temp > self._max_temp:
            self._status = "error"
            raise ValueError(
                f"Temperature {temp}¬∞C out of range "
                f"[{self._min_temp}, {self._max_temp}]"
            )
        self._value = temp
        self._status = "ok"
    
    @property
    def status(self):
        """Current sensor status."""
        return self._status
    
    @property
    def is_valid(self):
        """Check if sensor has valid reading."""
        return self._status == "ok" and self._value is not None

# Create and use sensor
sensor = TemperatureSensor("TEMP_001")
print(f"Sensor: {sensor.sensor_id}")
print(f"Initial status: {sensor.status}")

# Valid reading
sensor.value = 25.5
print(f"\nAfter reading 25.5¬∞C:")
print(f"Value: {sensor.value}¬∞C")
print(f"Status: {sensor.status}")
print(f"Is valid: {sensor.is_valid}")

# Invalid reading
try:
    sensor.value = 150
except ValueError as e:
    print(f"\nError: {e}")
    print(f"Status: {sensor.status}")
````

## Part 9: Composition


Composition is a design principle where objects contain other objects as attributes. It represents a "has-a" relationship (a Car HAS an Engine) rather than an "is-a" relationship (which is inheritance, covered in Week 11).

### Basic Composition Example

**Figure 9.1: Composition - Car and Engine**

````python
# Composition: Car HAS an Engine
class Engine:
    """Engine class for a vehicle."""
    
    def __init__(self, horsepower, fuel_type):
        self.horsepower = horsepower
        self.fuel_type = fuel_type
        self.is_running = False
    
    def start(self):
        """Start the engine."""
        self.is_running = True
        return f"Engine started ({self.horsepower}HP, {self.fuel_type})"
    
    def stop(self):
        """Stop the engine."""
        self.is_running = False
        return "Engine stopped"

class Car:
    """Car class that contains an Engine."""
    
    def __init__(self, brand, model, engine):
        self.brand = brand
        self.model = model
        self.engine = engine  # Composition: Car HAS an Engine
        self.speed = 0
    
    def start(self):
        """Start the car."""
        return self.engine.start()
    
    def stop(self):
        """Stop the car."""
        self.speed = 0
        return self.engine.stop()
    
    def accelerate(self, amount):
        """Accelerate the car."""
        if not self.engine.is_running:
            return "Start the engine first!"
        self.speed += amount
        return f"Speed: {self.speed} km/h"
    
    def __str__(self):
        return f"{self.brand} {self.model} ({self.engine.horsepower}HP)"

# Create engine and car
v8_engine = Engine(300, "gasoline")
car = Car("Ford", "Mustang", v8_engine)

print(car)
print(car.accelerate(20))  # Engine not running
print(car.start())
print(car.accelerate(30))
print(car.accelerate(40))
print(car.stop())
````

### Composition with Multiple Components

**Figure 9.2: Computer with Multiple Components**

````python
# Composition with multiple components
class CPU:
    """CPU component."""
    def __init__(self, model, cores, speed_ghz):
        self.model = model
        self.cores = cores
        self.speed_ghz = speed_ghz
    
    def __str__(self):
        return f"{self.model} ({self.cores} cores @ {self.speed_ghz}GHz)"

class Memory:
    """RAM memory component."""
    def __init__(self, size_gb, speed_mhz):
        self.size_gb = size_gb
        self.speed_mhz = speed_mhz
    
    def __str__(self):
        return f"{self.size_gb}GB @ {self.speed_mhz}MHz"

class Storage:
    """Storage component."""
    def __init__(self, size_gb, storage_type):
        self.size_gb = size_gb
        self.storage_type = storage_type
    
    def __str__(self):
        return f"{self.size_gb}GB {self.storage_type}"

class Computer:
    """Computer composed of multiple components."""
    
    def __init__(self, name, cpu, memory, storage):
        self.name = name
        self.cpu = cpu
        self.memory = memory
        self.storage = storage
        self.is_on = False
    
    def power_on(self):
        self.is_on = True
        return f"{self.name} powered on"
    
    def power_off(self):
        self.is_on = False
        return f"{self.name} powered off"
    
    def specs(self):
        """Display computer specifications."""
        print(f"=== {self.name} Specifications ===")
        print(f"CPU: {self.cpu}")
        print(f"Memory: {self.memory}")
        print(f"Storage: {self.storage}")
        print(f"Status: {'On' if self.is_on else 'Off'}")

# Create components
cpu = CPU("Intel i7-12700K", 12, 3.6)
ram = Memory(32, 3200)
ssd = Storage(1000, "NVMe SSD")

# Compose computer
my_pc = Computer("Workstation", cpu, ram, ssd)
my_pc.specs()
print()
print(my_pc.power_on())
my_pc.specs()
````

#### Engineering Example: Sensor Station

**Figure 9.3: Sensor Station with Composition**

````python
# Engineering: Sensor station using composition
class Sensor:
    """Individual sensor component."""
    
    def __init__(self, sensor_id, sensor_type, unit):
        self.sensor_id = sensor_id
        self.sensor_type = sensor_type
        self.unit = unit
        self.value = None
    
    def read(self, value):
        """Record a reading."""
        self.value = value
    
    def __str__(self):
        if self.value is not None:
            return f"{self.sensor_id}: {self.value} {self.unit}"
        return f"{self.sensor_id}: No reading"

class DataLogger:
    """Data logging component."""
    
    def __init__(self, name):
        self.name = name
        self.readings = []
    
    def log(self, sensor):
        """Log a sensor reading."""
        if sensor.value is not None:
            self.readings.append({
                'id': sensor.sensor_id,
                'type': sensor.sensor_type,
                'value': sensor.value,
                'unit': sensor.unit
            })
    
    def get_readings(self):
        return self.readings
    
    def clear(self):
        self.readings = []

class SensorStation:
    """Weather station composed of multiple sensors and a logger."""
    
    def __init__(self, station_id):
        self.station_id = station_id
        self.sensors = []  # Composition: has multiple sensors
        self.logger = DataLogger(f"{station_id}_logger")  # Composition: has a logger
    
    def add_sensor(self, sensor):
        """Add a sensor to the station."""
        self.sensors.append(sensor)
    
    def read_all(self, values):
        """Read values into all sensors and log them."""
        for sensor, value in zip(self.sensors, values):
            sensor.read(value)
            self.logger.log(sensor)
    
    def status(self):
        """Display station status."""
        print(f"=== Station {self.station_id} ===")
        for sensor in self.sensors:
            print(f"  {sensor}")
        print(f"  Total logged readings: {len(self.logger.readings)}")

# Create sensor station
station = SensorStation("WEATHER_01")

# Add sensors (composition)
station.add_sensor(Sensor("TEMP", "temperature", "¬∞C"))
station.add_sensor(Sensor("HUMI", "humidity", "%"))
station.add_sensor(Sensor("PRES", "pressure", "hPa"))
station.add_sensor(Sensor("WIND", "wind_speed", "km/h"))

# Simulate readings
station.read_all([25.5, 65.0, 1013.25, 12.5])
station.read_all([26.0, 63.5, 1013.50, 15.0])
station.read_all([25.8, 64.0, 1013.40, 14.0])

station.status()
print(f"\nLast 2 logged readings:")
for reading in station.logger.get_readings()[-2:]:
    print(f"  {reading}")
````

> üí° **Note:** Composition promotes code reuse and flexibility. You can easily swap components (e.g., use a different logger or sensor) without changing the main class. It also makes testing easier‚Äîyou can test components independently.

### Inventory Management System

**Figure 10.1: Product Inventory System**

````python
# Complete example: Inventory Management System
class Product:
    """A product in the inventory."""
    
    product_count = 0  # Class attribute
    
    def __init__(self, product_id, name, price, quantity=0):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity
        Product.product_count += 1
    
    @property
    def total_value(self):
        """Calculate total value of this product."""
        return self.price * self.quantity
    
    def restock(self, amount):
        """Add stock."""
        if amount > 0:
            self.quantity += amount
            return True
        return False
    
    def sell(self, amount):
        """Sell stock."""
        if amount > 0 and amount <= self.quantity:
            self.quantity -= amount
            return True
        return False
    
    def __str__(self):
        return f"{self.product_id}: {self.name} (${self.price}) x {self.quantity}"

class Inventory:
    """Inventory management system."""
    
    def __init__(self, name):
        self.name = name
        self.products = {}
    
    def add_product(self, product):
        """Add a product to inventory."""
        self.products[product.product_id] = product
    
    def remove_product(self, product_id):
        """Remove a product from inventory."""
        if product_id in self.products:
            del self.products[product_id]
            return True
        return False
    
    def get_product(self, product_id):
        """Get a product by ID."""
        return self.products.get(product_id)
    
    @property
    def total_value(self):
        """Calculate total inventory value."""
        return sum(p.total_value for p in self.products.values())
    
    def low_stock_alert(self, threshold=5):
        """Return products with low stock."""
        return [p for p in self.products.values() if p.quantity <= threshold]
    
    def report(self):
        """Generate inventory report."""
        print(f"=== {self.name} Inventory Report ===")
        print(f"Total products: {len(self.products)}")
        print(f"Total value: ${self.total_value:.2f}")
        print("\nProducts:")
        for product in self.products.values():
            print(f"  {product}")
        
        low_stock = self.low_stock_alert()
        if low_stock:
            print(f"\n‚ö†Ô∏è Low stock alert ({len(low_stock)} items):")
            for p in low_stock:
                print(f"  - {p.name}: {p.quantity} remaining")

# Create inventory and products
store = Inventory("Electronics Store")

store.add_product(Product("P001", "Laptop", 999.99, 10))
store.add_product(Product("P002", "Mouse", 29.99, 50))
store.add_product(Product("P003", "Keyboard", 79.99, 3))
store.add_product(Product("P004", "Monitor", 399.99, 8))

# Sell some items
store.get_product("P001").sell(3)
store.get_product("P002").sell(45)

# Generate report
store.report()
````

### Robot Control System

**Figure 10.2: Simple Robot Simulation**

````python
# Engineering: Simple robot control system
class Motor:
    """Motor component for robot movement."""
    
    def __init__(self, name, max_speed):
        self.name = name
        self.max_speed = max_speed
        self.current_speed = 0
    
    def set_speed(self, speed):
        """Set motor speed with limits."""
        self.current_speed = max(0, min(speed, self.max_speed))
    
    def stop(self):
        """Stop the motor."""
        self.current_speed = 0
    
    def __str__(self):
        return f"{self.name}: {self.current_speed}/{self.max_speed}"

class Gripper:
    """Gripper component for object manipulation."""
    
    def __init__(self):
        self.is_open = True
        self.holding = None
    
    def open(self):
        """Open the gripper."""
        self.is_open = True
        dropped = self.holding
        self.holding = None
        return dropped
    
    def close(self):
        """Close the gripper."""
        self.is_open = False
    
    def grab(self, item):
        """Grab an item."""
        if self.is_open:
            self.close()
            self.holding = item
            return True
        return False
    
    def __str__(self):
        state = "open" if self.is_open else "closed"
        return f"Gripper: {state}, holding: {self.holding}"

class Robot:
    """A robot composed of motors and a gripper."""
    
    def __init__(self, name):
        self.name = name
        self.left_motor = Motor("left", 100)
        self.right_motor = Motor("right", 100)
        self.gripper = Gripper()
        self.position = [0, 0]
    
    def move_forward(self, speed):
        """Move robot forward."""
        self.left_motor.set_speed(speed)
        self.right_motor.set_speed(speed)
        self.position[1] += 1
        return f"Moving forward at speed {speed}"
    
    def turn_left(self):
        """Turn robot left."""
        self.left_motor.set_speed(0)
        self.right_motor.set_speed(50)
        return "Turning left"
    
    def turn_right(self):
        """Turn robot right."""
        self.left_motor.set_speed(50)
        self.right_motor.set_speed(0)
        return "Turning right"
    
    def stop(self):
        """Stop the robot."""
        self.left_motor.stop()
        self.right_motor.stop()
        return "Robot stopped"
    
    def pick_up(self, item):
        """Pick up an item."""
        if self.gripper.grab(item):
            return f"Picked up: {item}"
        return "Gripper is not open"
    
    def put_down(self):
        """Put down the held item."""
        item = self.gripper.open()
        if item:
            return f"Put down: {item}"
        return "Nothing to put down"
    
    def status(self):
        """Display robot status."""
        print(f"=== Robot {self.name} ===")
        print(f"Position: {self.position}")
        print(f"Motors: {self.left_motor}, {self.right_motor}")
        print(f"{self.gripper}")

# Create and control robot
robot = Robot("R2D2")
robot.status()

print("\n--- Executing Commands ---")
print(robot.move_forward(50))
print(robot.pick_up("red cube"))
print(robot.move_forward(75))
print(robot.turn_left())
print(robot.put_down())
print(robot.stop())

print()
robot.status()
````

## ‚ùå Common Mistakes to Avoid


These are the most frequent errors students make with OOP. Study them before the exercises!

**Forgetting `self` in method definitions**

`def get_name():` ‚Üí `TypeError: takes 0 positional arguments but 1 was given`. Instance methods always need `self` as the first parameter: `def get_name(self):`

**Forgetting `self.` when accessing attributes**

`return name` inside a method looks for a local variable, not the attribute. Always use `return self.name` to access instance data.

**Mutable class attributes shared between instances**

`class Dog: tricks = []` ‚Äî ALL Dog instances share the same list! Define mutable attributes in `__init__`: `self.tricks = []`

**Confusing `__str__` and `__repr__`**

`__str__` is for human-readable display (`print()`). `__repr__` is for developers/debugging and should ideally be a valid Python expression. If only one is defined, implement `__repr__` ‚Äî it's used as fallback.

**Not calling `__init__` properly**

`s = Sensor` assigns the class itself, not an instance! You need parentheses: `s = Sensor("T01", "temperature")` to call the constructor.

## üî¨ Case Study: Sensor Data Logger (Part 3 of 6)


Continuing from Week 9's error-handling pipeline, we now wrap everything in a proper **OOP design** with `Sensor` and `SensorLogger` classes.

**Goal:** Refactor the procedural pipeline into classes ‚Äî a `Sensor` that holds readings and computes stats, and a `SensorLogger` that manages multiple sensors and generates reports.

**What's new:** `__init__`, instance methods, `__str__`/`__repr__`, properties, composition (Logger contains Sensors).

**Case Study 3 ‚Äî Sensor Data Logger: OOP Design**

````python
# === CASE STUDY Part 3: Sensor Data Logger ‚Äî OOP ===
# Refactor from functions + dicts ‚Üí classes

class Sensor:
    """Represents a single sensor with readings."""
    
    def __init__(self, sensor_id, sensor_type, unit):
        self.sensor_id = sensor_id
        self.sensor_type = sensor_type
        self.unit = unit
        self._readings = []     # Private: list of (timestamp, value)
        self._active = True
    
    @property
    def reading_count(self):
        return len(self._readings)
    
    @property
    def is_active(self):
        return self._active
    
    def add_reading(self, timestamp, value):
        """Add a validated reading."""
        if not self._active:
            raise RuntimeError(f"Sensor {self.sensor_id} is deactivated")
        self._readings.append((timestamp, value))
    
    def get_stats(self):
        """Calculate min, max, average statistics."""
        if not self._readings:
            return {"min": None, "max": None, "avg": None, "count": 0}
        values = [v for _, v in self._readings]
        return {
            "min": min(values), "max": max(values),
            "avg": sum(values) / len(values),
            "count": len(values)
        }
    
    def deactivate(self):
        self._active = False
    
    def __str__(self):
        stats = self.get_stats()
        status = "ACTIVE" if self._active else "INACTIVE"
        return (f"Sensor({self.sensor_id}, {self.sensor_type}, "
                f"{stats['count']} readings, {status})")
    
    def __repr__(self):
        return f"Sensor('{self.sensor_id}', '{self.sensor_type}', '{self.unit}')"


class SensorLogger:
    """Manages multiple sensors ‚Äî composition pattern."""
    
    def __init__(self, name):
        self.name = name
        self._sensors = {}   # id ‚Üí Sensor
        self._error_log = []
    
    def register_sensor(self, sensor):
        """Add a sensor to the logger."""
        self._sensors[sensor.sensor_id] = sensor
        print(f"  üì° Registered: {sensor}")
    
    def log_reading(self, sensor_id, timestamp, value):
        """Log a reading with error handling."""
        try:
            sensor = self._sensors[sensor_id]
            sensor.add_reading(timestamp, float(value))
        except KeyError:
            self._error_log.append(f"Unknown sensor: {sensor_id}")
        except (ValueError, RuntimeError) as e:
            self._error_log.append(f"{sensor_id}: {e}")
    
    def generate_report(self):
        """Generate a summary report."""
        lines = [f"{'='*50}",
                 f"  üìä {self.name} ‚Äî Status Report",
                 f"{'='*50}"]
        for sid, sensor in self._sensors.items():
            stats = sensor.get_stats()
            lines.append(f"\n  [{sid}] {sensor.sensor_type.upper()}")
            if stats["count"] > 0:
                lines.append(f"    Range: {stats['min']:.1f} ‚Äì {stats['max']:.1f} {sensor.unit}")
                lines.append(f"    Average: {stats['avg']:.2f} {sensor.unit}")
                lines.append(f"    Readings: {stats['count']}")
            else:
                lines.append(f"    No readings yet")
        if self._error_log:
            lines.append(f"\n  ‚ö†Ô∏è Errors: {len(self._error_log)}")
            for err in self._error_log:
                lines.append(f"    ‚Ä¢ {err}")
        lines.append(f"\n{'='*50}")
        return "\n".join(lines)


# --- Using our OOP Design ---
print("üèóÔ∏è Setting up Sensor Logger...\n")
logger = SensorLogger("Lab Station Alpha")

# Register sensors
t_sensor = Sensor("T01", "temperature", "¬∞C")
h_sensor = Sensor("H01", "humidity", "%")
logger.register_sensor(t_sensor)
logger.register_sensor(h_sensor)

# Log readings (including some bad data)
print("\nüìù Logging readings...")
test_data = [
    ("T01", "2024-01-15 08:00", "22.5"),
    ("T01", "2024-01-15 08:05", "23.1"),
    ("H01", "2024-01-15 08:00", "45.2"),
    ("H01", "2024-01-15 08:05", "N/A"),    # Bad data!
    ("X99", "2024-01-15 08:10", "10.0"),    # Unknown sensor!
    ("T01", "2024-01-15 08:10", "22.9"),
]

for sid, ts, val in test_data:
    logger.log_reading(sid, ts, val)

# Generate report
print(f"\n{logger.generate_report()}")
print("\n‚úÖ Clean OOP design: Sensor holds data, Logger orchestrates!")
print("üîú Next week: Inheritance for specialized sensor types!")
````
