## 1.1 What Are Comparison Operators?

Comparison operators are like asking questions:
- "Is 5 equal to 5?" ‚Üí Yes (True)
- "Is 5 greater than 10?" ‚Üí No (False)

The answer is always **True** or **False** - these are called **Boolean** values.

| Operator | Question it Asks | Example | Answer |
|----------|-----------------|---------|--------|
| `==` | Are they equal? | `5 == 5` | `True` |
| `!=` | Are they different? | `5 != 3` | `True` |
| `>` | Is left bigger? | `5 > 3` | `True` |
| `<` | Is left smaller? | `5 < 3` | `False` |
| `>=` | Is left bigger or equal? | `5 >= 5` | `True` |
| `<=` | Is left smaller or equal? | `5 <= 3` | `False` |

## ‚ö†Ô∏è IMPORTANT: `=` vs `==`

This is the **#1 mistake** beginners make!

| Symbol | Name | What it does | Example |
|--------|------|--------------|--------|
| `=` | Assignment | Puts a value into a variable | `x = 5` (x now holds 5) |
| `==` | Comparison | Asks "are these equal?" | `x == 5` (returns True or False) |

Think of it this way:
- One `=` is like **writing** on a sticky note
- Two `==` is like **asking a question**

````python
# Let's see the difference between = and ==

# = ASSIGNS a value (puts something in a box)
x = 5
print(f"x is now: {x}")

# == COMPARES values (asks a question)
result = x == 5
print(f"Is x equal to 5? {result}")

result = x == 10
print(f"Is x equal to 10? {result}")
````

## 1.2 Equal To (`==`) and Not Equal To (`!=`)

````python
# Equal to (==) - Are these the same?
print("=== Equal To (==) ===")
print(f"5 == 5: {5 == 5}")      # Yes, same number
print(f"5 == 3: {5 == 3}")      # No, different numbers
print(f"'Ali' == 'Ali': {'Ali' == 'Ali'}")  # Yes, same text
````

````python
# Not equal to (!=) - Are these different?
print("=== Not Equal To (!=) ===")
print(f"5 != 3: {5 != 3}")      # Yes, they're different
print(f"5 != 5: {5 != 5}")      # No, they're the same
print(f"'cat' != 'dog': {'cat' != 'dog'}")  # Yes, different
````

## 1.3 Greater Than and Less Than

````python
# Greater than (>) - Is the left side bigger?
print("=== Greater Than (>) ===")
print(f"10 > 5: {10 > 5}")   # Yes, 10 is bigger
print(f"5 > 10: {5 > 10}")   # No, 5 is smaller
print(f"5 > 5: {5 > 5}")     # No, they're equal (not greater)
````

````python
# Less than (<) - Is the left side smaller?
print("=== Less Than (<) ===")
print(f"3 < 7: {3 < 7}")     # Yes, 3 is smaller
print(f"7 < 3: {7 < 3}")     # No, 7 is bigger
print(f"5 < 5: {5 < 5}")     # No, they're equal (not less)
````

````python
# Greater than OR equal to (>=)
print("=== Greater Than or Equal (>=) ===")
print(f"5 >= 5: {5 >= 5}")   # Yes, 5 equals 5
print(f"6 >= 5: {6 >= 5}")   # Yes, 6 is greater than 5
print(f"4 >= 5: {4 >= 5}")   # No, 4 is less than 5
````

````python
# Less than OR equal to (<=)
print("=== Less Than or Equal (<=) ===")
print(f"5 <= 5: {5 <= 5}")   # Yes, 5 equals 5
print(f"4 <= 5: {4 <= 5}")   # Yes, 4 is less than 5
print(f"6 <= 5: {6 <= 5}")   # No, 6 is greater than 5
````

## 1.7 ‚ö†Ô∏è Be Careful with Float Comparisons!

Due to how computers store decimal numbers, comparing floats can be tricky!

````python
# This might surprise you!
a = 0.1 + 0.2
b = 0.3

print(f"0.1 + 0.2 = {a}")
print(f"0.3 = {b}")
print(f"Are they equal? {a == b}")  # Surprise! They're NOT equal!
````

````python
# Better way to compare floats: use a tolerance
a = 0.1 + 0.2
b = 0.3
tolerance = 0.0001  # How close is "close enough"?

# Check if the difference is very small
difference = abs(a - b)  # abs() gives absolute value
print(f"Difference: {difference}")
print(f"Are they approximately equal? {difference < tolerance}")
````

## 2.1 What is an `if` Statement?

An `if` statement lets your program **make decisions**. It's like saying:

> "**IF** this is true, **THEN** do this."

```python
if condition:
    # This code runs ONLY if condition is True
    # Notice the 4-space indent!
```

**Real-life example:**
- IF it's raining ‚Üí bring an umbrella
- IF you're hungry ‚Üí eat food
- IF battery is low ‚Üí charge phone

## ‚ö†Ô∏è CRITICAL: Python Indentation Rules

Python uses **indentation** (spaces at the beginning of lines) to group code!

```python
if condition:
    print("This is inside the if")    # 4 spaces = inside
    print("This too")                  # 4 spaces = inside
print("This is outside the if")        # 0 spaces = outside
```

**Rules:**
1. Use **4 spaces** for each indent level (or press Tab)
2. All code in the same block must have the **same** indentation
3. Don't mix spaces and tabs!
4. Wrong indentation = `IndentationError`

````python
# Basic if statement
age = 20

if age >= 18:
    print("You are an adult!")  # This line is INSIDE the if (4 spaces)

print("Program continues...")    # This line is OUTSIDE the if (0 spaces)
````

````python
# What happens when condition is False?
age = 15

if age >= 18:
    print("You are an adult!")  # This does NOT print (condition is False)

print("Program continues...")    # This ALWAYS prints (outside the if)
````

````python
# Multiple lines inside an if block
score = 85

if score >= 60:
    print("üéâ Congratulations!")
    print("You passed the exam!")
    print(f"Your score: {score}")
    print("Keep up the good work!")

print("--- End of Program ---")
````

## 4.1 What is `elif`?

`elif` is short for "else if". Use it when you have **more than two** possible outcomes.

```python
if condition1:
    # Runs if condition1 is True
elif condition2:
    # Runs if condition1 is False AND condition2 is True
elif condition3:
    # Runs if condition1 and condition2 are False AND condition3 is True
else:
    # Runs if ALL conditions are False
```

**Think of it like a flowchart:**
- Check first condition ‚Üí if True, done!
- If False, check second condition ‚Üí if True, done!
- If False, check third condition ‚Üí and so on...
- If nothing matches, run `else` (optional)

````python
# Grade Calculator
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} ‚Üí Grade: {grade}")
````

````python
# Let's test multiple scores!
scores = [95, 85, 75, 65, 45]

print("Grade Report")
print("="*20)

for score in scores:
    if score >= 90:
        grade = "A"
    elif score >= 80:
        grade = "B"
    elif score >= 70:
        grade = "C"
    elif score >= 60:
        grade = "D"
    else:
        grade = "F"
    print(f"{score}: {grade}")
````

````python
# Positive, Negative, or Zero
number = 0

if number > 0:
    print(f"{number} is POSITIVE")
elif number < 0:
    print(f"{number} is NEGATIVE")
else:
    print(f"{number} is ZERO")
````

## 5.1 What Are Logical Operators?

## 5.2 The `and` Operator

**Both conditions must be True** for the result to be True.

Think of it like: "I need my keys **AND** my wallet to leave"

| A | B | A and B |
|---|---|--------|
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |

````python
# AND truth table
print("=== AND Truth Table ===")
print(f"True and True:   {True and True}")
print(f"True and False:  {True and False}")
print(f"False and True:  {False and True}")
print(f"False and False: {False and False}")
````

## 6.1 What Are Nested Conditionals?

You can put an `if` statement **inside** another `if` statement!

```python
if condition1:
    # First check
    if condition2:
        # Second check (only if first passed)
```

**Real-life example:**
- IF you have a ticket ‚Üí IF you're old enough ‚Üí you can enter

````python
# Event entry check
age = 25
has_ticket = True

if age >= 18:
    print("‚úì You are old enough.")
    if has_ticket:
        print("‚úì You have a ticket.")
        print("üéâ Welcome to the event!")
    else:
        print("‚úó Please buy a ticket first.")
else:
    print("‚úó Sorry, adults only.")
````

````python
# ATM Withdrawal System
balance = 500
withdrawal = 200
pin_correct = True

if pin_correct:
    print("‚úì PIN verified")
    if withdrawal <= balance:
        balance = balance - withdrawal
        print(f"üíµ Dispensing ${withdrawal}")
        print(f"üìä New balance: ${balance}")
    else:
        print("‚úó Insufficient funds!")
        print(f"Your balance: ${balance}")
else:
    print("‚úó Incorrect PIN!")
````

````python
# Number classification with deep nesting
number = 15

if number >= 0:
    print(f"{number} is non-negative")
    if number == 0:
        print("It's zero")
    else:
        print("It's positive")
        if number % 2 == 0:
            print("It's even")
        else:
            print("It's odd")
else:
    print(f"{number} is negative")
````

````python
# Engineering: Multi-level safety check
power_on = True
safety_engaged = True
temperature = 45

if power_on:
    print("Power: ON ‚úì")
    if safety_engaged:
        print("Safety: ENGAGED ‚úì")
        if temperature < 60:
            print("Temperature: OK ‚úì")
            print(">>> System ready to operate!")
        else:
            print("Temperature: TOO HIGH ‚úó")
            print(">>> Cool down required.")
    else:
        print("Safety: DISENGAGED ‚úó")
        print(">>> Cannot operate. Engage safety first.")
else:
    print("Power: OFF")
    print(">>> Turn on power to begin.")
````

## ‚ö†Ô∏è Warning: Don't Over-Nest!

Too many nested levels make code hard to read!

**If you have more than 3 levels, consider:**
1. Using logical operators (`and`, `or`)
2. Breaking into separate functions
3. Restructuring your logic

## 7.1 What is the Ternary Operator?

A shorthand way to write simple `if-else` statements on **one line**.

```python
value = result_if_true if condition else result_if_false
```

````python
# Traditional if-else
age = 20
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
print(f"Traditional: {status}")

# Ternary operator - same thing in one line!
status = "Adult" if age >= 18 else "Minor"
print(f"Ternary: {status}")
````

````python
# Even or Odd
number = 7
result = "Even" if number % 2 == 0 else "Odd"
print(f"{number} is {result}")
````

````python
# Pass or Fail
score = 75
result = "PASS ‚úì" if score >= 60 else "FAIL ‚úó"
print(f"Score {score}: {result}")
````

````python
# Maximum of two numbers
a = 15
b = 23
maximum = a if a > b else b
print(f"Max of {a} and {b}: {maximum}")
````

````python
# Absolute value (without abs() function)
number = -7
absolute = number if number >= 0 else -number
print(f"Absolute value of {number}: {absolute}")
````

````python
# Using ternary inside f-strings
is_online = True
print(f"Server is {'online üü¢' if is_online else 'offline üî¥'}")

# Plural handling
count = 1
print(f"You have {count} {'item' if count == 1 else 'items'}")

count = 5
print(f"You have {count} {'item' if count == 1 else 'items'}")
````

## 8.1 What is the Match Statement?

Python 3.10 introduced `match` for pattern matching - like a cleaner version of many `elif` statements.

```python
match value:
    case pattern1:
        # action1
    case pattern2:
        # action2
    case _:  # default (like else)
        # default action
```

````python
# Day of week using match
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6 | 7:  # Multiple values with |
        print("Weekend! üéâ")
    case _:  # Default case (like else)
        print("Invalid day")
````

````python
# HTTP Status Codes
status_code = 404

match status_code:
    case 200:
        message = "OK ‚úì"
    case 201:
        message = "Created"
    case 400:
        message = "Bad Request"
    case 404:
        message = "Not Found üîç"
    case 500:
        message = "Internal Server Error"
    case _:
        message = "Unknown Status"

print(f"Status {status_code}: {message}")
````

````python
# Simple calculator using match
operation = "+"
a = 10
b = 5

match operation:
    case "+":
        result = a + b
    case "-":
        result = a - b
    case "*":
        result = a * b
    case "/":
        result = a / b if b != 0 else "Error: Division by zero"
    case _:
        result = "Unknown operation"

print(f"{a} {operation} {b} = {result}")
````

# üêõ Part 9: Common Mistakes to Avoid

Let's look at the most common errors beginners make!

## Mistake 1: Using `=` instead of `==`

````python
# WRONG (this causes an error or unexpected behavior)
# if x = 5:  # SyntaxError!

# CORRECT
x = 5
if x == 5:
    print("x is 5")
````

## Mistake 2: Missing Colon `:`

````python
# WRONG
# if score >= 60   # Missing colon!
#     print("Passed")

# CORRECT
score = 75
if score >= 60:  # Don't forget the colon!
    print("Passed")
````

## Mistake 3: Wrong Indentation

````python
# WRONG
# if age >= 18:
# print("Adult")  # IndentationError - not indented!

# CORRECT
age = 20
if age >= 18:
    print("Adult")  # 4 spaces indent
````

## Mistake 4: Using `&&` or `||` instead of `and` / `or`

````python
# WRONG (from other languages like Java, C++, JavaScript)
# if x > 5 && y < 10:  # SyntaxError!

# CORRECT (Python uses words)
x = 7
y = 5
if x > 5 and y < 10:
    print("Both conditions true!")
````

## Mistake 5: Floating Point Comparison

````python
# DANGEROUS - might fail!
result = 0.1 + 0.2
if result == 0.3:
    print("Equal!")  # This might NOT print!
else:
    print("Not equal (surprise!)")

# BETTER - use tolerance
if abs(result - 0.3) < 0.0001:
    print("Approximately equal ‚úì")
````
